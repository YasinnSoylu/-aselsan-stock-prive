```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# AAPL hisse senedinin fiyat verilerini yfinance ile al
aapl = yf.Ticker("AAPL")
data = aapl.history(period="1y")

# SMA(20) hesapla
sma_20 = data["Close"].rolling(window=20).mean()

# Güncel fiyatı al ve SMA(20) ile karşılaştır
current_price = data["Close"].iloc[-1]
sma_20_current = sma_20.iloc[-1]

if current_price > sma_20_current:
    signal = "ALIM"
    color = "green"
elif current_price < sma_20_current:
    signal = "SATIM"
    color = "red"
else:
    signal = "BEKLE"
    color = "blue"

# Grafik olarak göster
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(data["Close"], label="Fiyat")
ax.plot(sma_20, label="SMA(20)")
ax.scatter(data.index[data['Close'] > sma_20], data['Close'][data['Close'] > sma_20], color=color, marker='^', label=signal)
ax.scatter(data.index[data['Close'] < sma_20], data['Close'][data['Close'] < sma_20], color="RED", marker='v', label=signal)
ax.legend(loc="upper left")
ax.set_title(f"AAPL Al/Sat Sinyali: {signal}")
plt.show()
```
